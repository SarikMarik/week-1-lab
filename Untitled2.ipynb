PROPLEM 3

```python
scores=[45, 67, 82, 59]
def pass_fail(score):
  if score>=60:
    return "Pass"
  else:
    return "Fail"
for score in scores:
  result=pass_fail(score)
  print(f"Input:{score} -> Output: {result}")
```

PROBLEM 4

```python
sizes=[500, 800, 1000, 1200]
prices=[50, 80, 100, 130]
for i in range(len(sizes)):
  print("Input:", sizes[i], "square meter -> Output:", prices[i], "thousand dollars")

is_increasing=True
for i in range(1, len(prices)):
  if prices[i]<=prices[i-1]:
    is_increasing= False

if is_increasing:
  print("Positive trend")
else:
  print("Negative trend")
```
