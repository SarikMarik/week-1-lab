PROBLEM 4

```python
sizes=[500, 800, 1000, 1200]
prices=[50, 80, 100, 130]
for i in range(len(sizes)):
  print("Input:", sizes[i], "square meter -> Output:", prices[i], "thousand dollars")

is_increasing=True
for i in range(1, len(prices)):
  if prices[i]<=prices[i-1]:
    is_increasing= False

if is_increasing:
  print("Positive trend")
else:
  print("Negative trend")
```

```python
study_hourse=[2, 4, 6, 8]
scores=[55, 80, 100, 130]
#Calculate increases
score_increases=[]
for i in range(1, len(scores)):
  increase=scores[i]-scores[i-1]
  score_increases.append(increase)
#Average increase for every 2 hours
avg_increase=sum(score_increases)/len(score_increases)
# Predict score for 10 hours
predicted_score=scores[-1]+avg_increase

# Print predicted score
print("Predicted score for 10 hourse of study:", predicted_score)
```
